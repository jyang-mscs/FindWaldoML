With CNN, it was not realize that CNN is a image recognition focused model and to achieve the goal of "finding Waldo" takes more than just image recogition. The correct classification of the issue would be a object identification type problem.

Attempts to use the CNN model with other approaches to achieve the purpose of the goal, like the sliding window approach did not showed any success. Model 1 showed to the be ideal model out of 3 total models, having an AUC or 0.9973 and a validation loss of 0.1062 still could not find Waldo.

At the same time, consultations with Professor Yu (Zeyun Yu) who has a focus on image processing for other models similar to CNN. Professor Yu recommended using the YOLO model. Without much background o YOLO and it was also not covered as part of this semester, online sources was used to further expand the knowledge of YOLO.

The following reference link below takes to the website/blog post that explains the usuage on implementation of the YOLO model. Following the instructions, a similar model was recreated for the purpose of "finding Waldo".
[Source](https://neptune.ai/blog/object-detection-with-yolo-hands-on-tutorial)

```python
# Step 1: Install dependencies and clone YOLOv5 repository
import os

# Clone YOLOv5 repository
if not os.path.exists('yolov5'):
    os.system("git clone https://github.com/ultralytics/yolov5")
    os.system("pip install -r yolov5/requirements.txt")

import torch
import yaml
import shutil
```

```python
# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
# Step 2: Define paths for dataset and configurations

# final_project_YOLO.ipynb
dataset_path = "/Users/jeffyang/CS715-Final/dataset_yolo"
train_images = os.path.join(dataset_path, "images/train")
val_images = os.path.join(dataset_path, "images/val")
train_labels = os.path.join(dataset_path, "labels/train")
val_labels = os.path.join(dataset_path, "labels/val")
```

```python
# YAML Configuration File for YOLOv5
data_config = {
    'path': dataset_path,         # Root dataset path
    'train': 'images/train',      # Training images directory
    'val': 'images/val',          # Validation images directory
    'nc': 2,                      # Number of classes (1 for "Waldo")
    'names': ['waldo', 'notwaldo']            # Class names
}

# Save YAML file
yaml_file = 'waldo.yaml'
with open(yaml_file, 'w') as file:
    yaml.dump(data_config, file, default_flow_style=False)

print("Dataset YAML configuration saved!")
```

```python
# Step 3: Start YOLOv5 Training
print("Starting YOLOv5 Training...")

# Define YOLO training parameters
yolo_params = {
    'img_size': 640,             # Image size for training
    'batch_size': 16,            # Batch size (adjust based on GPU memory)
    'epochs': 100,               # Number of training epochs
    'data': yaml_file,           # Path to dataset YAML file
    'weights': 'yolov5s.pt',     # Pretrained weights for transfer learning
    'name': 'waldo_detector',    # Experiment name
    'device': device             # Use GPU or CPU
}
```

```python
# Start training
os.system(
    f"python yolov5/train.py --img {yolo_params['img_size']} "
    f"--batch {yolo_params['batch_size']} --epochs {yolo_params['epochs']} "
    f"--data {yolo_params['data']} --weights {yolo_params['weights']} "
    f"--name {yolo_params['name']} --device {yolo_params['device']}"
)

print("Training completed!")
```

```python
# Step 4: Evaluate the Model
print("Evaluating the trained model...")

trained_weights_path = os.path.abspath(f"yolov5/runs/train/{yolo_params['name']}/weights/best.pt")

if not os.path.exists(trained_weights_path):
    print(f"Error: {trained_weights_path} does not exist.")
else:
    print(f"Found: {trained_weights_path}")
    
```

```python
os.system(
    f"python yolov5/val.py --weights {trained_weights_path} "
    f"--data {yaml_file} --img {yolo_params['img_size']}"
)
```

```python


print("Evaluation complete!")

# Step 5: Run Inference
print("Running inference with the trained model...")

import cv2
from matplotlib import pyplot as plt

# Load trained model
model = torch.hub.load('ultralytics/yolov5', 'custom', path=trained_weights_path, force_reload=True)

# Test image path
test_image_path = "data/Hey-Waldo-master/original-images/3.jpg"  # Change this to the image you want to test

# Load test image
image = cv2.imread(test_image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Perform inference
results = model(image_rgb)

# Display results
results.show()

# Save results
results.save()

print("Inference completed. Results saved to 'runs/detect/exp'.")
```
